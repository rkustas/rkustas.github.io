```python
import pandas as pd
import numpy as np
import matplotlib as mp
import requests
import os
import json
```

```python
# Testing a request and getting all necessary url strings
test_url = "https://games.crossfit.com/competitions/api/v1/competitions/open/2018/leaderboards?division=1&region=0&scaled=0&sort=0&occupation=0&page=1"
years_url = "https://games.crossfit.com/competitions/api/v1/competitions/open/"
years_url2 = "/leaderboards?division=1&region=0&scaled=0&sort=0&occupation=0&page="
request = requests.get(test_url)
test_loads = json.loads(request.text)
test_loads
```

```python
# Years to evaluate
years = ["2018","2019"]
```

```python
# Evaluate the total number of pages in each year
for year in years:
    full_url = years_url +year+years_url2+ "1"
    print(full_url)
    resp = requests.get(full_url)
    data2 = json.loads(resp.text)
    total_pages = np.arange(1,3001)
    print(total_pages)
    for page in total_pages:
        print(f"Page {page}")
```

```python
# Custom errors to pass
class Error(Exception):
    pass
class ErrortoPass(Error):
    pass
```

## Code for Men events

```python
# Putting the desired request information into a dataframe, used 3000 pages of data in each year
comp_info = []
columns = ["Competitor_id","Competitor_name","Event_num","Event_rank","Event_score","Event_score_display","Event_year","Event_time","Event_judge"]
for year in years:
    full_url = years_url +year+years_url2+ "1"
#     print(full_url)
    resp = requests.get(full_url).json()
#     print(resp)
#     print(resp['pagination']['totalPages'])
    total_pages = np.arange(1,3001)
#     print(total_pages)
    for page in total_pages:
        print(f"Page {page}")
#         print(f"{years_url}{year}{years_url2}{page}")
        full_url = years_url + str(year) + years_url2 + str(page)
#         print(full_url)
        response = requests.get(full_url).json()
        row = []
        try:
            for r in response['leaderboardRows']:
    #             print(r)
                c_id = r['entrant']['competitorId']
                c_n = r['entrant']['competitorName']
                for s in r['scores']:
#                         print(s)
                        e_num = s['ordinal']
                        e_rank = s['rank']
                        e_score = s['score']
                        e_s_dis = s['scoreDisplay']
                        e_year = year
                        try:
                            e_time = s['time']
                        except KeyError:
                            e_time = "No time"
                        try:
                            e_judge = s['judge']
                        except KeyError:
                            e_judge = "No judge"
                        row = [c_id,c_n,e_num,e_rank,e_score,e_s_dis,e_year,e_time,e_judge]
                        comp_info.append(row)
        except ErrortoPass as e:
                print(e)
```

```python
# Printing sample of DF
df = pd.DataFrame(comp_info,columns=columns)
df.head()
```

```python
# Checking the shape of the df
df.shape
```

```python
# Getting unique competitor information
competitorinfo = []
columnscomp = ["Competitor_id","Competitor_name","first_name","last_name","status","gender","region_name","affiliate_id","affiliate_name","age","height","weight"]
for year in years:
    full_url = years_url +year+years_url2+ "1"
    resp = requests.get(full_url).json()
    total_pages = np.arange(1,3001)
    for page in total_pages:
        print(f"Page {page}")
        full_url = years_url + str(year) + years_url2 + str(page)
        response = requests.get(full_url).json()
        row = []
        try:
            for r in response['leaderboardRows']:
                c_id = r['entrant']['competitorId']
                c_n = r['entrant']['competitorName']
                f_n = r['entrant']['firstName']
                l_n = r['entrant']['lastName']
                s = r['entrant']['status']
                g = r['entrant']['gender']
                try:
                    r_n = r['entrant']['regionName']
                except KeyError:
                    r_n = "No Region"
                a_id = r['entrant']['affiliateId']
                a_n = r['entrant']['affiliateName']
                a = r['entrant']['age']
                h = r['entrant']['height']
                w = r['entrant']['weight']
                row = [c_id,c_n,f_n,l_n,s,g,r_n,a_id,a_n,a,h,w]
                competitorinfo.append(row)
        except ErrortoPass as e:
            print(e)
```

```python
# Competitor Info DF
comp_df = pd.DataFrame(competitorinfo,columns=columnscomp)
comp_df
```

```python
# Concatenate women and men competitor info 
frames_comp = [df_comp_men,df_comp_wo]
comb_comp_df = pd.concat(frames_comp)
comb_comp_df.head()
```

```python
# Checking types of comp df
comb_comp_df.dtypes
```

```python
# Split height column and add back to original df
new_height = comb_comp_df['height'].str.split(" ", n = 1, expand = True).replace(r'^\s*$', 0, regex=True)
comb_comp_df['height'] = new_height[0]
comb_comp_df['height_units'] = new_height[1]
comb_comp_df
```

```python
# Split weight column and add back to original df
new_weight = comb_comp_df['weight'].str.split(" ", n = 1, expand = True).replace(r'^\s*$', 0, regex=True)
comb_comp_df['weight'] = new_weight[0]
comb_comp_df['weight_units'] = new_weight[1]
comb_comp_df
```

```python
comb_comp_df.dtypes
```

```python
# Convert height and weight to int64
comb_comp_df['weight'] = comb_comp_df['weight'].astype(float)
comb_comp_df['height'] = comb_comp_df['height'].astype(float)
comb_comp_df.dtypes
```

```python
comb_comp_df.head()
```

```python
# Convert height to inches and weight to lbs
comb_comp_df['height'] = np.round(np.where(comb_comp_df['height_units'] == "cm", comb_comp_df['height'] / 2.54, comb_comp_df['height']), 1)
comb_comp_df['weight'] = np.round(np.where(comb_comp_df['weight_units'] == "kg", comb_comp_df['height'] * 2.2, comb_comp_df['weight']), 1)
comb_comp_df
```

```python
# Drop units column

comb_comp_df = comb_comp_df.drop(columns=['height_units', 'weight_units'])
comb_comp_df.head()
```
