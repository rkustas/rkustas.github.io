## Basic ML

### Age vs. overall Score

```python
X= leaderboard['age'].values.reshape(-1,1)
y = leaderboard['overallScore'].values.reshape(-1,1)

print("Shape: ", X.shape, y.shape)
```

```python
plt.scatter(X, y,c='r')
plt.xlabel("age")
plt.ylabel("overallScore")
plt.figure(figsize=(15,15),facecolor='r',dpi=100)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")
```

```python
df3['Rank_quantiles'] = pd.qcut(df3['overallRank'], q=10,labels=labels)
df3.head()
```

## ML model
* Goal is to whether you make it into the top 10 on the leaderboard based on height, weight, age and event score

```python
X= df3[['height', 'weight', 'age', 's_score']].values
y = df3[['Rank_quantiles']].values.reshape(-1, 1)
print("Shape: ", X.shape, y.shape)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X_scaler = StandardScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train.shape
# y_train = to_categorical(y_train)
# y_train[:10]
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=4))
model.add(Dense(units=10, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
yt1 = y_train.flatten() - 1
yt = to_categorical(yt1, num_classes=10)
yt
```

```python
yt2 = y_test.flatten() -1
yt_test = to_categorical(yt2, num_classes=10)
yt_test
```

```python
model.fit(
    X_train_scaled,
    yt,
    epochs=5000,
    shuffle=True,
    verbose=2
)
```

```python
model.summary()
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, yt_test, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
test = np.expand_dims(X_test_scaled[1], axis=0)
test.shape
```

```python
# Make a prediction. The result should be 5 - STANDING
print(f"Predicted class: {model.predict_classes(test)}")
```

## Kmeans

```python
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, yt)
    train_score = knn.score(X_train_scaled,yt)
    test_score = knn.score(X_test_scaled, yt_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 20, 2), train_scores, marker='o')
plt.plot(range(1, 20, 2), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_scaled, yt)
print('k=1 Test Acc: %.3f' % knn.score(X_test_scaled, yt_test))
```
