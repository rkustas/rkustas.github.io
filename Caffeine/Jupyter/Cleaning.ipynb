```python
import pandas as pd
import numpy as np
import matplotlib as mp
import os
import stats
from itertools import chain
```

```python
# CWD
os.getcwd()
```

```python
# Setting path and iterating through it to push all CSVs to a dictionary
root = "C:\\Users\\Ryan\\OneDrive - ogg.com\\DU_class\\Caffeine-Project\\Caffeine\\CSVs"

ddict = {}
for file in os.listdir(root):
    if file.endswith(".csv"):
        name = os.path.splitext(file)[0]
        ddict[name] = pd.read_csv(os.path.join(root, file))
print(ddict)
```

```python
# Renaming dicts to dfs
alldrink = ddict['All_drink_info']
caff_int = ddict['Caffeine_intensity']
food = ddict['Food']
gum = ddict['Gum']
workout = ddict['Workout Supps']
```

## Viewing dfs to see if they need cleaning

```python
alldrink
```

```python
caff_int
caff_int['caff_str'] = caff_int['caff_str'].str.replace(' and sugar.','')
df_obj = caff_int.select_dtypes(['object'])
caff_int[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
caff_int
```

```python
food['caff_per_mg']= food['caff_per_mg'].astype(int)
food.head()
```

```python
# Checking types
food.dtypes
```

```python
gum.head()
```

```python
# Cleaning
gum.head()
gum['Caff_per_piece'] = gum['Caff_per_piece'].map(lambda x: x.rstrip('mg'))
gum['Price_per_pack'] = gum['Price_per_pack'].map(lambda x: x.lstrip('$'))
gum['Price_per_pack'] = gum['Price_per_pack'].map(lambda x: x.lstrip('\u20AC'))
gum['Price_per_pack'] = gum['Price_per_pack'].map(lambda x: x.rstrip('ct'))
gum['Price_per_pack'] = gum['Price_per_pack'].map(lambda x: x.rstrip('or less'))
gum['Price_per_pack'] = gum['Price_per_pack'].map(lambda x: x.strip('vari'))
```

```python
# Remove columns without desired data
fil = gum['Caff_per_piece']!= 'none'
gum = gum[fil]
fil2 = gum['Price_per_pack']!=''
gum = gum[fil2]
gum.head()
```

```python
# Checking types
gum.dtypes
gum['Caff_per_piece'] = gum['Caff_per_piece'].astype(float)
gum['Price_per_pack'] = gum['Price_per_pack'].astype(float)
gum,gum.dtypes
```

```python
gum.dtypes
```

```python
# Further gum cleaning

new_gum = (gum.set_index(['Gum_name', 'Caff_per_piece','Price_per_pack','Page_url'])
   .apply(lambda x: x.str.split(',').explode())
   .reset_index())
```

```python
df_obj = new_gum.select_dtypes(['object'])
new_gum[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
new_gum
```

```python
workout.head()
workout[['caff','serving_size']] = workout.caff_per_serv.str.split("/",expand=True)
workout = workout.drop('caff_per_serv',axis=1)
```

```python
workout['caff'] = workout['caff'].astype(int)
workout
```

```python
workout['serving_size']= workout['serving_size'].str.replace('*','')
df_obj = workout.select_dtypes(['object'])
workout[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
workout
```

```python
# Export to CSV for db loading
alldrink.to_csv("CSVs/All_drink_info.csv",index=False)
caff_int.to_csv("CSVs/Caffeine_intensity.csv",index=False)
workout.to_csv("CSVs/Workout Supps.csv",index=False)
food.to_csv("CSVs/Food.csv",index=False)
new_gum.to_csv("CSVs/Gum.csv",index=False)
```
